# Desafio 2: Duas formas de selecionar os dados por bairro (consultar os métodos na documentação do pandas)

## Opção 1: Dicionário

```python
# Quais são os bairros?

bairros = dados["Bairro"].value_counts().index

# Uma possível solução é armazenar em um dicionário que segue a estrutura em que 
# as chaves do dicionário são os bairros e os valores são os DataFrames com 
# valores referentes ao seu respectivo bairro 

dados_por_bairro = {}

for bairro in bairros:
    mask = dados["Bairro"] == bairro
    df_tmp = dados[mask]
    dados_por_bairro[bairro] = df_tmp

dados_por_bairro
```

```python
# Assim, ao acessar a chave "Vila Mariana" tem-se 

dados_por_bairro["Vila Mariana"]
```

## Opção 2: método pd.grouby()

```python
# A variável agrupado armazena os dados agrupados por bairro

agrupado = dados.groupby("Bairro")
```

## Opção 3: método set_index()

```python
dados.set_index("Bairro")
```

```python
# Retorna todos os valores em que o Bairro é Vila Mariana

dados.set_index("Bairro").loc["Vila Mariana"]
```

## Opção 4: método query()

```python
# Retorna todos os valores em que o Bairro é Vila Mariana

dados.query("Bairro == 'Vila Mariana'")
```

# Desafio 5: Descobrir quais são os bairros que não tem nome de Rua.

```python
dados.info()
```

Utilizando o método info() nota-se que existem valores nulos para a 'Rua'. Pode-se verificar quais valores da series Rua são nulos utilizando o método isnull(), como realizado a seguir

```python
dados["Rua"].isnull()
```

O resultado é uma série booleana que servirá como máscara para filtrar as linhas do DataFrame que apresentam elementos do atributo "Rua" que são nulos/NaN.

```python
sum(dados["Rua"].isnull())
```

Neste caso existem 3434 valores de "Rua" que são nulos. A seguir é realizada a filtragem dos dados. O métodos unique() neste caso retorna os valores únicos dos bairros cujo foi apresentado o valor de "Rua" nulo ao menos uma vez.

```python
dados["Bairro"][dados["Rua"].isnull()].unique()
```

A quantidade de dados únicos pode ser aferiada como mostrado abaixo.

```python
len(dados["Bairro"][dados["Rua"].isnull()].unique())
```
