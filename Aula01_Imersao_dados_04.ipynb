# Desafio 1: Realizar a média da metragem para cada um dos bairros

```python
# Quais são os bairros?

dados["Bairro"].value_counts().index
```

```python
# Filtrar os apartamentos por bairro e extrair a média da metragem. Pode-se utilizar o método groupby do pandas.

agrupado = dados.groupby("Bairro")

agrupado["Metragem"].mean()
```

```python
# Ou também pode-se fazer assim

agrupado.mean()["Metragem"]
```
